```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
data1 = pd.read_csv('1cpu100000sim.txt', sep=",", header=None)
data1.columns = ['core', 'runtime', 'comment']
sns.set(rc={'figure.figsize':(25, 7)})
sns.barplot(x='core', y='runtime', data=data1.sort_values(by=['runtime'], ascending=False)).set(title='Monte Carlo Simulation with 100.000 times on 50 assets')
plt.xlabel("Number of Core")
plt.ylabel("Runtime (hours)")
plt.show()
```

```python
data2 = pd.read_csv('vectorization100000sim.txt', sep=",", header=None)
data2.columns = ['core', 'runtime', 'comment']
#perf.core =  ["1cpu1core", "1cpu20core", "1cpu40core", "1cpu80core", "1cpu100core", "1cpu128core", "1cpu50task2cpu"]
#print(perf)
sns.set(rc={'figure.figsize':(20, 5)})
sns.barplot(x='core', y='runtime', data=data2.sort_values(by=['runtime'], ascending=False)).set(title='Monte Carlo Simulation with 100.000 times on 50 assets')
plt.xlabel("Number of Core")
plt.ylabel("Runtime (hours)")
plt.show()
```

```python

concatenated = pd.concat([data1.assign(optimization='Before Vectorization'), data2.assign(optimization='After Vectorization')])
sns.set(rc={'figure.figsize':(25, 7)})
sns.color_palette("Paired")
sns.barplot(x='core', y='runtime', hue='optimization', data=concatenated.sort_values(by=['runtime'], ascending=False), palette="Paired").set(title='Monte Carlo Simulation with 100.000 times on 50 assets')
plt.xlabel("Number of Core")
plt.ylabel("Runtime (Hours)")
plt.show()
```

```python
# specify plot layouts with different width using subplots()
f, axs = plt.subplots(1, 2,
   figsize = (25, 5),
   sharey = True,
   gridspec_kw = dict(width_ratios = [0.5, 0.5]))

# make scatterplot with legends
sns.barplot(data = data1.sort_values(by=['runtime'], ascending=False),
   x = "core",
   y = "runtime",
   ax = axs[0],
)
axs[0].set_xlabel("Number of Core")
axs[0].set_ylabel("Runtime (hours)")
# make densityplot with kdeplot without legends
sns.barplot(data = data2.sort_values(by=['runtime'], ascending=False),
   x = "core",
   y = "runtime",
   ax = axs[1])
axs[1].set_xlabel("Number of Core")
axs[1].set_ylabel("Runtime (hours)")
f.tight_layout()
```

```python
data1 = pd.read_csv('vectorization100000sim.txt', sep=",", header=None)
data1.columns = ['core', 'runtime', 'comment']
data1 = data1.drop([5, 6])

data2 = pd.read_csv('2cpunodes.txt', sep=",", header=None)
data2.columns = ['core', 'runtime', 'comment']
data2 = data2.drop([5, 6])

concatenated = pd.concat([data1.assign(nodes='1cpu node'), data2.assign(nodes='2cpu node')])
sns.set(rc={'figure.figsize':(25, 5)})
sns.color_palette("Paired")
sns.barplot(x='core', y='runtime', hue='nodes', data=concatenated.sort_values(by=['runtime'], ascending=False), palette="Paired").set(title='Monte Carlo Simulation with 100.000 times on 50 assets')
plt.xlabel("Number of Core")
plt.ylabel("Runtime (Hours)")
plt.show()
```
